```python
import copy
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_fruit_imgs():
    """
    Carrega as imagens de frutas guardados no diretório frutas
    Retorna as imagens em um array
    """
    return [plt.imread(f_i) for f_i in get_fruit_files()]

def get_fruit_files():
    """
    Carrega o dataset de imagens de frutas no diretório frutas
    Retorna os nomes de arquivos
    """

    files = [os.path.join('frutas', file_i)
             for file_i in os.listdir('frutas')
             if '.jpg' in file_i]
    return files
```

```python
files = get_fruit_files()
```

```python
# lê a primeira imagem do array (lista de frutas). O primeiro item do arranjo é sempre zero.
img1 = plt.imread(files[0])

# lê a oitava imagem do array (lista de frutas)
img2 = plt.imread(files[7])
```

```python
# plota as imagens lidas do array
plt.figure()
plt.imshow(img1)
plt.figure()
plt.imshow(img2)
```

```python
# imprime em texto os valores da camada 'red' da img1 
# (img1[:,:,X] o valor X representa a camada: 0 - vermelho, 1 - verde, 2 - azul)
# a organização das funções print() serve somente para apresentar os dados como uma matriz 

for j in img1[:,:,0]: 
    for i in j: 
        print(i, end=" ")
    print("")
```

```python
rimg=copy.deepcopy(img1);
rimg[:,:,1]=0
rimg[:,:,2]=0
gimg=copy.deepcopy(img1);
gimg[:,:,0]=0
gimg[:,:,2]=0
bimg=copy.deepcopy(img1);
bimg[:,:,0]=0
bimg[:,:,1]=0
plt.figure()
plt.imshow(rimg)
plt.figure()
plt.imshow(gimg)
plt.figure()
plt.imshow(bimg)
```

```python
# Faz a cópia das imagens em novas variáveis e preenche de zeros os valores das demais camadas, mantendo
# somente os dados da camada desejada

rimg=copy.deepcopy(img2);
rimg[:,:,1]=0
rimg[:,:,2]=0
gimg=copy.deepcopy(img2);
gimg[:,:,0]=0
gimg[:,:,2]=0
bimg=copy.deepcopy(img2);
bimg[:,:,0]=0
bimg[:,:,1]=0

# plota as imagens de cada camada
plt.figure()
plt.imshow(rimg)
plt.figure()
plt.imshow(gimg)
plt.figure()
plt.imshow(bimg)
```
