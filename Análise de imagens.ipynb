```python
import copy
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_celeb_files():
    """
    Carrega o dataset de imagens de celebridades no diretório
    img_align_celeba
    Retorna os nomes de arquivos
    """

    files = [os.path.join('img_align_celeba', file_i)
             for file_i in os.listdir('img_align_celeba')
             if '.jpg' in file_i]
    return files

def get_celeb_imgs():
    """
    Carrega as imagens de celebridades guardados no diretório
    Retorna as imagens em um array
    """
    return [plt.imread(f_i) for f_i in get_celeb_files()]
```

```python
files = get_celeb_files()
```

```python
img = plt.imread(files[0])
```

```python
plt.imshow(img)
```

```python
for j in img[:,:,0]: 
    for i in j: 
        print(i, end=" ")
    print("")
```

```python
rimg=copy.deepcopy(img);
rimg[:,:,1]=0
rimg[:,:,2]=0
plt.imshow(rimg)
```

```python
bimg=copy.deepcopy(img);
bimg[:,:,0]=0
bimg[:,:,1]=0
plt.imshow(bimg)
```

```python
imagens = get_celeb_imgs()
```

```python
plt.imshow(imagens[2])
```

```python
dados = np.array(imagens)
```

```python
dados.shape
```

```python
mean_img = np.mean(dados,axis=0)
```

```python
max_img = np.max(dados,axis=0)
```

```python
min_img = np.min(dados,axis=0)
```

```python
std_img = np.std(dados,axis=0)
```

```python
plt.imshow(mean_img.astype(np.uint8))
```

```python
plt.imshow(max_img.astype(np.uint8))
```

```python
plt.imshow(min_img.astype(np.uint8))
```

```python
plt.imshow(np.mean(std_img,axis=2).astype(np.uint8))
plt.colorbar()
```
